```python
import pandas as pd
from sqlalchemy import create_engine
```

# Load consumer_complaints Dataframe by Product = Mortage and Sub-Product = VA

```python
csv_file = "./Resources/consumer_complaints.csv"
consumer_data_df = pd.read_csv(csv_file)
consumer_data_df.head()
```

```python
consumer_data_df.columns
```

```python
clean_consumer_data_df = consumer_data_df[
['date_received', 'product', 'sub_product', 'issue', 'company',
       'state', 'zipcode', 'submitted_via', 'date_sent_to_company', 'company_response_to_consumer',
       'timely_response', 'consumer_disputed?', 'complaint_id']
].copy()
clean_consumer_data_df.head()
```

```python
clean_consumer_data_df.info()
```

```python
clean_consumer_data_df['product'].unique()
```

```python
clean_consumer_data_df['product'].value_counts()
```

```python
#Mortage product
clean_consumer_data_df[clean_consumer_data_df['product'] == 'Mortgage']['sub_product'].unique()
```

```python
mortgage_df = clean_consumer_data_df[clean_consumer_data_df['product'] == 'Mortgage']
```

```python
mortgage_df.head()
```

```python
# VA mortage
va_mortgage_df = mortgage_df[mortgage_df['sub_product'] == 'VA mortgage']
```

```python
va_mortgage_df.head()
```

```python
va_mortgage_df.head()
```

```python
# Rename the column headers
va_mortgage_df = va_mortgage_df.rename(columns={"consumer_disputed?": "consumer_disputed"})
va_mortgage_df.head()
```

# Load States DataFrame

```python
#state data frame
state_csv_file = "./Resources/states.csv"
state_data_df = pd.read_csv(state_csv_file)
state_data_df.head()
```

# Load Vet By Age Group DataFrame

```python
#Vet by age group
vet_by_age_csv_file = "./Resources/vets_by_age_group.csv"
vet_by_age_data_df = pd.read_csv(vet_by_age_csv_file)

vet_by_age_data_df.head()
```

# Load Vet By Race DataFrame

```python
#Vet by age race
vet_by_race_csv_file = "./Resources/vets_by_race.csv"
vet_by_race_data_df = pd.read_csv(vet_by_race_csv_file)
vet_by_race_data_df.head()
```

# Connect to postgress database

```python
connection_string = "postgres:postgres@localhost:5432/etl_project"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
va_mortgage_df.to_sql(name='consumer_complaints', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from consumer_complaints', con=engine).head()
```

```python
state_data_df.to_sql(name='states', con=engine, index=False, if_exists='append')
```

```python
pd.read_sql_query('select * from states', con=engine).head()
```

```python
vet_by_age_data_df.to_sql(name='vets_by_age_group', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from vets_by_age_group', con=engine).head()
```

```python
pd.read_sql_query('select * from vets_by_race', con=engine).head()
```

```python
vet_by_race_data_df.to_sql(name='vets_by_race', con=engine, if_exists='append', index=False)
```
